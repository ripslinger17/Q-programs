```python
import cirq
import numpy as np
import cirq_web.bloch_sphere as bloch_sphere
import matplotlib.pyplot as plt
```

```python
# Creating Qubit
q0, q1, q2 = cirq.LineQubit.range(0,3)
```

```python
# Creating Circuit
circuit = cirq.Circuit()
```

```python
# Adding Gates
x_gate1 = cirq.X(q0)
x_gate2 = cirq.X(q0)
```

```python
# Adding to Gates to Circuit
circuit.append(x_gate1)
#xcircuit.append(x_gate2)
print(circuit)
```

```python
measure = cirq.measure(q0, key="key1")
circuit.append(measure)
print(circuit)
```

```python
# Creating Simulator
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)
q0_result = result.measurements["key1"]
print(q0_result)
```

```python
# Plotting Histogram
cirq.plot_state_histogram(result, plt.subplot())
plt.show()
```
