```python
import cirq
import matplotlib.pyplot as plt
import cirq_web.bloch_sphere as BlochSphere
```

```python
# Creating Qubits
q0 = cirq.LineQubit(0)
```

```python
# Creating Circuit
circuit = cirq.Circuit()
```

```python
# Adding Gates
x_gate1 = cirq.X(q0)
x_gate2 = cirq.X(q0)
z_gate1 = cirq.Z(q0)
z_gate2 = cirq.Z(q0)
h_gate1 = cirq.H(q0)
h_gate2 = cirq.H(q0)
h_gate3 = cirq.H(q0)
```

```python
# Adding Gates on Circuit
circuit.append(h_gate1)
circuit.append(x_gate1)
circuit.append(z_gate1)
circuit.append(x_gate2)
circuit.append(h_gate2)
circuit.append(z_gate2)
circuit.append(h_gate3)
print(circuit)
```

```python
# measurement
measure = cirq.measure(q0, key = "key1")
circuit.append(measure)
print(circuit)
```

```python
# Simulator
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)
q0_result = result.measurements["key1"]
print(q0_result)
```

```python
# Output
cirq.plot_state_histogram(result, plt.subplot())
plt.show()
```
