# 2.3 Measure the outcome of the Problem 2.2 in values and graph

```python
import cirq
import matplotlib.pyplot as plt
```

```python
# Creating Qubits and Circuit
q0, q1, q2 = cirq.LineQubit.range(0,3)
circuit = cirq.Circuit()
```

```python
# Adding Gates
H_gate = cirq.H(q0)
I_gate1 = cirq.I(q1)
I_gate2 = cirq.I(q2)
```

```python
# Adding Gates to Circuit
circuit.append(H_gate)
circuit.append(I_gate1)
circuit.append(I_gate2)
print(circuit)
```

```python
# Adding Measurement
measure = cirq.measure(q0, key="key1")
circuit.append(measure)
print(circuit)
```

```python
# Simulating Circuit
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)
measurement_result = result.measurements["key1"]
print(measurement_result)
```

```python
# Plotting Histogram
cirq.plot_state_histogram(result, plt.subplot())
plt.show()
```
